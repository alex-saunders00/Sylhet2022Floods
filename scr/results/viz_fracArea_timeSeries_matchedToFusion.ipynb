## PART 1: Import the already created FFA data for the different sources

```python
# Import the FIA from the matched rasters
fracFloodAreaMatched = pd.read_csv(rootPath/'Table/SylhetFracFloodedArea/FIA_matchedToFusion_sylhetDist.csv')
fracFloodAreaMatched.ImageDate = [str(datetime.strptime(str(item), '%Y%m%d').date()) for item in fracFloodAreaMatched.ImageDate]
fracFloodAreaMatched.index = pd.to_datetime(fracFloodAreaMatched.ImageDate)
```

```python
# Import the precipitation data from csv file, extracted via the code "precip_time_series.ipynb"
inputPath = rootPath/'Table/SylhetPrecip'
precip_daily_all = pd.read_csv(inputPath/'precip_daily_all.csv', index_col='date')
precip_daily_all.index = pd.to_datetime(precip_daily_all.index)
```

```python
# Prep the data from the df for plotting
GFMPlot = fracFloodAreaMatched[fracFloodAreaMatched.MapSource=='GFM']
DevriesPlot = fracFloodAreaMatched[fracFloodAreaMatched.MapSource=='Devries']
IMPACTPlot = fracFloodAreaMatched[fracFloodAreaMatched.MapSource=='IMPACT']
UNOSATPlot = fracFloodAreaMatched[fracFloodAreaMatched.MapSource=='UNOSAT']
FusionPlot = fracFloodAreaMatched[fracFloodAreaMatched.MapSource=='Fusion']
GFDFloodedPlot = fracFloodAreaMatched[np.logical_and(fracFloodAreaMatched.MapSource=='GFD',fracFloodAreaMatched.ImageBand=='flooded')]
GFDMaxPlot = fracFloodAreaMatched[np.logical_and(fracFloodAreaMatched.MapSource=='GFD',fracFloodAreaMatched.ImageBand=='max_img')]
```

## PART 2: Visualize the fractional flooded area time series for Sylhet District
Plot time series of FFA, along with precipitation data and gauge level for the same period

```python
 # Plot a time series of fractional flooded area with subplot of precipitation, with FFA for just Sylhet, without confusion matrix metrics

# Create the plot
fig, ax = plt.subplots(2, sharex=True, figsize=(11, 5), height_ratios=[1,2], layout='constrained') #11, 7
fig.suptitle('Sylhet District Precipitation and Fractional Inundated Area', fontsize=16)

colors_FFA = ['xkcd:dark blue', 'xkcd:turquoise', 'xkcd:sky blue', 'aqua', 'xkcd:tomato', 'xkcd:orange']#, 'xkcd:fuchsia']
labels_FFA = ['Thomas et al.', 'Paul & Ganju', 'GFM', 'UNOSAT', 'Fusion', 'GFD']#, 'GFD MaxImg']

# First subplot - plot the precip data
ax[0].plot(precip_daily_all.index, precip_daily_all['dist_median'], color = 'xkcd:red', linestyle = '-', label = 'Sylhet District')
ax[0].fill_between(precip_daily_all.index, precip_daily_all['dist_pctile_95'], precip_daily_all['dist_pctile_5'], color = 'xkcd:red', alpha = 0.2, label = None) #'Sylhet District (95th/5th %iles)'
ax[0].plot(precip_daily_all.index, precip_daily_all['hydro_dist_median'], color = 'xkcd:salmon', linestyle = '-', label = 'Watershed')
ax[0].fill_between(precip_daily_all.index, precip_daily_all['hydro_dist_pctile_95'], precip_daily_all['hydro_dist_pctile_5'], color = 'xkcd:salmon', alpha = 0.2, label = None) #'Watershed District (95th/5th %iles)'

# Set the yaxis, gridlines and limits
ymajor_tick = MultipleLocator(400)
yminor_tick = MultipleLocator(200)
ax[0].yaxis.set_major_locator(ymajor_tick)
ax[0].yaxis.set_minor_locator(yminor_tick)
ax[0].tick_params(axis='y', which='major', labelsize=14)
ax[0].set_ylabel('Daily \n precipitation (mm)', fontsize=14)
ax[0].grid(True, axis='x', which='major', linestyle='-')
ax[0].grid(True, axis='x', which='minor', linestyle=':')
ax[0].grid(True, axis='y', which='major', linestyle='-')
ax[0].grid(True, axis='y', which='minor', linestyle=':')
ax[0].set_ylim(0, math.ceil(max(list(precip_daily_all.values.flatten()))/100)*100)

# Plot the legend
# ax[0].legend(loc='center', bbox_to_anchor=[0.5, 1.2], ncol=4, frameon=True, fontsize=12) #title='ROI' #upper right
ax[0].legend(loc='upper right', ncol=2, frameon=True, fontsize=12) #title='ROI' #upper right


# Second subplot - plot the FFA time series, for modified DeVries and GFM, also add UNOSAT as data points

# Plot modified Devries
ax[1].plot(DevriesPlot.index, DevriesPlot.MeanFIA, marker= 's', color = colors_FFA[0], markersize=6, markeredgecolor='black', linestyle = ':', label = labels_FFA[0])
ax[1].plot(IMPACTPlot.index, IMPACTPlot.MeanFIA, marker= 's', color = colors_FFA[1], markersize=6, markeredgecolor='black', linestyle = ':', label = labels_FFA[1])
ax[1].plot(GFMPlot.index, GFMPlot.MeanFIA, marker= 's', color = colors_FFA[2], markersize=6, markeredgecolor='black', linestyle = ':', label = labels_FFA[2])
ax[1].plot(UNOSATPlot.index, UNOSATPlot.MeanFIA, marker= 's', color = colors_FFA[3], markersize=7, markeredgecolor='black', linestyle = ':', label = labels_FFA[3])
ax[1].plot(FusionPlot.index, FusionPlot.MeanFIA, marker= 'o', color = colors_FFA[4], markersize=6, markeredgecolor='black', linestyle = ':', label = labels_FFA[4])
ax[1].plot(GFDFloodedPlot.index, GFDFloodedPlot.MeanFIA, marker= 'o', color = colors_FFA[5], markersize=6, markeredgecolor='black', linestyle = ':', label = labels_FFA[5])
# ax[1].plot(GFDMaxPlot.index, GFDMaxPlot.MeanFIA, marker= 'o', color = colors_FFA[6], markersize=6, markeredgecolor='black', linestyle = ':', label = labels_FFA[6])


# Format the ticks, gridlines and axis labels
ymajor_tick = MultipleLocator(0.2)
yminor_tick = MultipleLocator(0.1)
ax[1].yaxis.set_major_locator(ymajor_tick)
ax[1].yaxis.set_minor_locator(yminor_tick)
ax[1].tick_params(axis='y', which='major', labelsize=14)
ax[1].set_ylabel('Fractional \n Inundated Area', fontsize=14, wrap=True)
ax[1].grid(True, axis='x', which='major', linestyle='-')#, color='black')
ax[1].grid(True, axis='x', which='minor', linestyle=':')
ax[1].grid(True, axis='y', which='major', linestyle='-')
ax[1].grid(True, axis='y', which='minor', linestyle=':')

# Format the x and y limits
#plt.xlim(fracFloodArea.index[0], fracFloodArea.index[len(fracFloodArea)-1])
ax[1].set_xlim(DevriesPlot.index[0]-timedelta(days=1), DevriesPlot.index[len(DevriesPlot)-1]+timedelta(days=1))
ax[1].set_ylim(0, 0.7)

# Add legend
ax[1].legend(loc='center', bbox_to_anchor=[0.5, 1.1], ncol=6, frameon=True, fontsize=12) #upper right

# Format the time series date settings
# week_locator = mdates.DayLocator(interval=7)
# ax.xaxis.set_major_locator(week_locator) 
day_locator = mdates.DayLocator(0)
ax[1].xaxis.set_minor_locator(day_locator)
ax[1].set_xticks(fracFloodAreaDevries.index, fracFloodAreaDevries.index, fontsize=12)
date_formatter = mdates.DateFormatter("%Y-%m-%d")
ax[1].xaxis.set_major_formatter(date_formatter)


# Finish the plot
plt.setp(ax[1].get_xticklabels(), rotation=60, horizontalalignment='right')
#fig.autofmt_xdate()
# fig.tight_layout()
plt.show()
```
