# Sylhet Floods 2022
## precip_time_series.ipynb
This script gets precipitation estimates for the ROI and TOI associated to the June 2022 Sylhet flooding.

Precipitation estimates are obtained via GEE from the Global Precipitation Mission (NASA's GPM), as half-hourly precipitation. We compute daily precipitation rate over the TOI for Sylhet.

```python
# Initialize the GEE API, trigger the athentication and initialize the library
import ee
```

```python
# Imports
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
from datetime import timedelta
```

## PART 0: Load imports

```python
district = ee.FeatureCollection("projects/ee-nip/assets/ShapeFiles/Admin/Districts")
gpm = ee.ImageCollection("NASA/GPM_L3/IMERG_V06")
hydrosheds7 = ee.FeatureCollection("WWF/HydroSHEDS/v1/Basins/hybas_7")
```

## PART 1: Define the ROI

```python
# Define the ROI as the Upazilla (sub-district) of Sylhet
roi_dist = district.filter(ee.Filter.eq('NAME_2', 'Sylhet'))

# Further define an ROI using HydroSHEDS basins - this will help to capture some upstream influence when computing precip estimates
# Selected HydroSHEDS L7 as appropriate resolution cacthments
# Union all the HydroSHEDS L7 that intersect with the Sylhet district
roi_hydro = hydrosheds7.filter(ee.Filter.bounds(roi_dist)).union()
```

## PART 2: Define the TOI and filter the precipitation data

```python
# Define start and end dates
start_date = ee.Date('2022-05-01');
end_date = ee.Date('2022-09-01');
```

```python
# Define the band name for the half hourly precipitation in mm
band_name = 'precipitationCal'
```

```python
# Get the GPM [precipitation data for the TOI
precip_collection = gpm.filterDate(start_date, end_date).select(band_name) # filterBounds not needed as is a global tile
print(type(precip_collection)) # ImageCollection
```

## PART 3: Get and visualize daily precipitation

```python
# Create and apply a function to aggregate the precip to daily rate
ndays = ee.Date(end_date).difference(ee.Date(start_date), 'days')

def dayOffset(list_sequence):
    start = ee.Date(start_date).advance(list_sequence, 'days')
    end = start.advance(1, 'days')
    return precip_collection.filterDate(start, end).sum().set('system:time_start', start.millis())

precip_daily = ee.ImageCollection(ee.List.sequence(0, ndays.subtract(1)).map(dayOffset))
```

```python
# Create a function to extract and prep the daily precip median, 5% and 95% %iles for a specified ROI
def get_daily_precip_stats(precip_collection, roi):
    
    # Define the feature geometry for the reduceRegion
    feature_geometry = roi.first().geometry()
    
    # Create a function to reduceRegion the precipitation image, for the ROIs - median value
    def reduce_precip_median(image):
        reduced_dict = image.reduceRegion(reducer = ee.Reducer.median(), 
                                          geometry = feature_geometry,
                                         scale = 11000) # GPM resolution is 11 km
        return image.set(reduced_dict)
    # Create a function to reduceRegion the precipitation image, for the ROIs - 95%ile value
    def reduce_precip_95(image):
        reduced_dict = image.reduceRegion(reducer = ee.Reducer.percentile([95]), 
                                          geometry = feature_geometry,
                                         scale = 11000) # GPM resolution is 11 km
        return image.set(reduced_dict)
    # Create a function to reduceRegion the precipitation image, for the ROIs - 5%ile value
    def reduce_precip_5(image):
        reduced_dict = image.reduceRegion(reducer = ee.Reducer.percentile([5]), 
                                          geometry = feature_geometry,
                                         scale = 11000) # GPM resolution is 11 km
        return image.set(reduced_dict)
    
    # Map the reducer function over the colletion
    precip_median = precip_collection.map(reduce_precip_median)
    precip_95 = precip_collection.map(reduce_precip_95)
    precip_5 = precip_collection.map(reduce_precip_5)
 
    # Extract the info as separate numpy arrays
    dateTime = np.array(precip_median.aggregate_array('system:time_start').getInfo())
    median = np.array(precip_median.aggregate_array(band_name).getInfo())
    pctile_95 = np.array(precip_95.aggregate_array(band_name).getInfo())
    pctile_5 = np.array(precip_5.aggregate_array(band_name).getInfo())
       
    # Create a pandas dataframe combining the arrays
    precip_df = pd.DataFrame(data = [dateTime, median, pctile_95, pctile_5], index = ['dateTime', 'median', 'pctile_95', 'pctile_5']).T
    
    # Convert the date format and add a simplified date column
    precip_df['dateTime'] = pd.to_datetime(precip_df['dateTime'], unit='ms')
    precip_df['date'] = precip_df['dateTime'].dt.date
    
    # Set the date as the index
    precip_df = precip_df.set_index('date')
    
    # Return the precipitation dataframe
    return precip_df
```

```python
# Extract the daily precipitation estimates for each of the ROIs
precip_daily_dist = get_daily_precip_stats(precip_daily, roi_dist)
precip_daily_hydro = get_daily_precip_stats(precip_daily, roi_hydro)
```

```python
# Combine and export the daily precip data for all ROIs
precip_daily_to_merge = [precip_daily_dist, precip_daily_hydro]
precip_daily_all = pd.concat(precip_daily_to_merge, join='outer', axis=1) # can use concat since we know the dates are all the same
precip_daily_all.drop('dateTime',axis=1, inplace=True)
precip_daily_all.columns = ['dist_median', 'dist_pctile_95', 'dist_pctile_5', 
                            'hydro_dist_median', 'hydro_dist_pctile_95', 'hydro_dist_pctile_5']
```

### Visualize the daily precipitation time series using the median and 5%ile / 95%ile

```python
# Create a function to visualize the time series for a given ROI
def visualize_precip_timeseries(precip_df, color, roi_name):
    
    # Define the x and y values for plotting
    x = precip_df.index
    y = precip_df['median']
    y1 = precip_df['pctile_95']
    y2 = precip_df['pctile_5']

    # Use fill_between to plot the confidence intervals
    #ax = ax or plt.gca()
    plt.plot(x, y, color = color, label = roi_name + ' (median)')
    plt.fill_between(x, y1, y2, color = color, alpha = 0.2, label = roi_name + ' (95th/5th %iles)')
    #return ax
```

```python
# Define a list of the ROI names
roi_names = ['Sylhet District']
colors = ['red']

# Create the plot
fig, ax = plt.subplots(figsize=(18, 3)) #figsize=(18, 6)
bottom, top = 0, 0.88
left, right = 0, 1
fig.subplots_adjust(top=top, bottom=bottom, left=left, right=right, hspace=0.1, wspace=0.1)

# Add the time series of each ROI
visualize_precip_timeseries(precip_daily_dist, colors[0], roi_names[0])

# Format the ticks, gridlines and axis labels
xmajor_tick = MultipleLocator(7)
xminor_tick = MultipleLocator(1)
ax.xaxis.set_major_locator(xmajor_tick)
ax.xaxis.set_minor_locator(xminor_tick)
ax.tick_params(axis='x', which='major', labelsize=12) 
ymajor_tick = MultipleLocator(100)
yminor_tick = MultipleLocator(50)
ax.yaxis.set_major_locator(ymajor_tick)
ax.yaxis.set_minor_locator(yminor_tick)
ax.tick_params(axis='y', which='major', labelsize=12)
plt.ylabel('Daily precipitation (mm)', fontsize=16)
ax.grid(True, which='major', linestyle='-')
ax.grid(True, which='minor', linestyle=':')

# Format the x and y limits
plt.xlim(precip_daily_dist['dateTime'].min(), precip_daily_dist['dateTime'].max())
plt.ylim(bottom = 0)

# Add legend
plt.legend(loc='upper left', frameon=True, fontsize=14)

plt.show()
```
